IMportando dados_SUS

```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv'
gastos_hospitalares = pd.read_csv(url2, encoding='ISO-8859-1', sep=';',skiprows=3,skipfooter=10,engine='python')
```

```python
gastos_hospitalares.tail()
```

```python
# Remove espaços em branco do início e do fim de todos os nomes das colunas
gastos_hospitalares.columns = gastos_hospitalares.columns.str.strip()
```

```python
print(gastos_hospitalares.columns)
# Assuming 'Total' is the last column based on the head() output
total_column = gastos_hospitalares.columns[-1]
print(total_column)
gastos_hospitalares = gastos_hospitalares.drop(total_column, axis=1)
```

```python
gastos_hospitalares.tail()
```

```python
df_Dados_hospitalares = pd.melt(gastos_hospitalares, id_vars=['Unidade da Federação'], var_name='data', value_name='Valor')
```

```python
df_Dados_hospitalares.head()
```

```python
df_Dados_hospitalares.to_csv('df_Dados_hospitalares.csv',index=False)
```
