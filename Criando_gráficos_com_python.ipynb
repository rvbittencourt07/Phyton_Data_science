```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv')
df
```

```python
df.info()
```

```python
df.set_index('País',inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil
```

```python
brasil_dict = {'ano' : brasil.index.tolist(),'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

Desafio: Visualizando dados de vendas de diferentes lojas

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_lojas = pd.DataFrame(vendas_2022,index=lojas)
df_lojas
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.8,wspace=0.5)
fig.suptitle('Vendas de 2022 por loja')

axs[0,0].plot(df_lojas.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df_lojas.loc['B'])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df_lojas.loc['C'])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df_lojas.loc['C'])
axs[1,1].set_title('Loja C')

#Adicionar titulos eixo X e eixo Y

for ax in axs.flat:
  ax.set_xlabel('Mês')
  ax.set_ylabel('Número de vendas')

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total',ascending=True)
america_sul_sorted
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig,ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('America do Sul: Brasil foi o 4º país com mais imigrantes\npara o Canadá de 1980 a 2013',loc='left',fontsize=18)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 50, i, str(v), color='black', fontsize = 10, ha='left', va = 'center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both',which='both',length=0)

fig.savefig('imigrantes_america_sul.png',transparent=False,dpi=300,bbox_inches='tight')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10,y = top_10.index,x='Total',orient='h',palette = palette)
  ax.set_title('Paises com maior imigração apra o Canadá\n1980 a 2013',fontsize = 18,loc='left')
  ax.set_xlabel('Numero de imigrantes',fontsize = 14)
  ax.set_ylabel('')
  plt.show()
```
